```python
import pandas as pd
from samplesize import sampleSize
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def aggregate_other(df, by, count_name='count', other_thr=1):
    df_c = df.copy()
    df_c = df_c[[by,count_name]]
    total_count = df_c[count_name].sum()
    df_c['per'] = df_c[count_name].apply(lambda x: 100*x/total_count)

    other_df_c = df_c[df_c['per'] < other_thr].sum()
    other_df_c[by] = 'other'

    df_c = df_c.append(
        other_df_c, ignore_index=True
    ).sort_values('per',ascending=False).reset_index(drop=True)
    df_c = df_c[df_c['per'] >= other_thr]

    return df_c
```

```python
data = []
from urllib.parse import unquote
with open('completed.txt','r',encoding='utf-8') as f:
    for line in f:
        l = unquote(line).replace('\n','').split(',')
        data.append({
            'item':','.join(l[0:-1]),
            'n_classes': int(l[-1])
        })
completed_df = pd.DataFrame(data).fillna('')
completed_df

completed_df['n_classes'].apply(lambda x : 'Not empty' if x>0 else 'Empty')\
    .value_counts(dropna=False, normalize=True)\
    .plot.bar(rot=0)
```

```python
data = []
i = 0
with open('candidate_classes.txt','r') as f:
    for line in f:
        i = i + 1
        line = line.replace('\n','')
        line_s = line.split(',')
        if line_s[0] == '':
            #print(i)
            continue
        row = {
            'item' : line_s[0]
        }
        for c in (',' + ','.join(line_s[1:])).split(',http://'):
            if c[:c.rfind('/')+1] in row:
                row[c[:c.rfind('/')+1]].append(c[c.rfind('/')+1:])
            else:
                row[c[:c.rfind('/')+1]] = [c[c.rfind('/')+1:]]
        data.append(row)
candidate_classes_df = pd.DataFrame(data).fillna('')
candidate_classes_df = candidate_classes_df.drop('',axis=1)

classes = []
for l in candidate_classes_df.drop('item', axis=1).values.tolist():
    c = [item for sublist in l for item in sublist]
    classes.append(len(c) if c != [''] else 0)
candidate_classes_df['n_classes'] = classes

candidate_classes_df.to_csv('candidate_classes_df.csv', index=False)
candidate_classes_df
```

```python
candidate_classes_df['n_classes']\
    .transform(lambda x : 'Empty' if x==0 else 'Not Empty')\
    .value_counts(dropna=False, normalize=True)\
    .plot.bar(rot=0)
```

```python
prefixes_df = candidate_classes_df.drop(['item','n_classes'],axis=1).apply(lambda x: [len(xx) for xx in x]).sum()
prefixes_df = pd.DataFrame(prefixes_df).reset_index()
prefixes_df.columns = ['prefix','count']

fig, ax = plt.subplots(figsize=[5,5])
sns.barplot(data = aggregate_other(prefixes_df, by='prefix'), x = 'per', y ='prefix', ax=ax)
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 5 if width < 15 else width - 5 ,
            p.get_y()+p.get_height()/2. + 0.2,
            '{:1.2f}'.format(width),
            ha="center")
```

```python
prefixes_coverage_df = candidate_classes_df.drop(['item','n_classes'],axis=1)\
    .apply(lambda x: [len(xx)>0 for xx in x])\
    .sum()/candidate_classes_df.shape[0]*100
prefixes_coverage_df = pd.DataFrame(prefixes_coverage_df).reset_index()
prefixes_coverage_df.columns = ['prefix','coverage']

fig, ax = plt.subplots(figsize=[5,5])
sns.barplot(
    data = aggregate_other(prefixes_coverage_df, by='prefix', count_name='coverage'),
    x = 'coverage', y ='prefix', ax=ax
)
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 1 if width < 12 else width - 1 ,
            p.get_y()+p.get_height()/2. + 0.2,
            '{:1.2f}'.format(width),
            ha="center")
```

```python
def calculate_coverage_prefixes(prefixes):
    return candidate_classes_df.drop(['item','n_classes'],axis=1)\
    .apply(lambda x: [len(xx)>0 for xx in x])[prefixes]\
    .sum(axis=1).apply(lambda x:x>0).sum()/candidate_classes_df.shape[0]
    
calculate_coverage_prefixes(['dbpedia.org/ontology/','dbpedia.org/class/yago/'])
```

```python
prefixes_to_pair = ['www.w3.org/2002/07/', 'xmlns.com/foaf/0.1/', 'www.ontologydesignpatterns.org/ont/dul/',
       'www.wikidata.org/entity/', 'schema.org/', 'umbel.org/umbel/rc/',
       'dbpedia.org/class/yago/', 'www.w3.org/2003/01/geo/',
       'purl.org/ontology/bibo/', 'dbpedia.org/class/', 'vocab.org/frbr/',
       'dbpedia.org/class/yago/WikicatATSC-M/',
       'dbpedia.org/class/yago/WikicatVFL/',
       'www.ontologydesignpatterns.org/ont/']
base_coverage = calculate_coverage_prefixes(['dbpedia.org/ontology/'])
for prefix in prefixes_to_pair:
    print('{} : +{}%'.format(prefix,
        100*(calculate_coverage_prefixes(['dbpedia.org/ontology/',prefix]) - base_coverage)))
```

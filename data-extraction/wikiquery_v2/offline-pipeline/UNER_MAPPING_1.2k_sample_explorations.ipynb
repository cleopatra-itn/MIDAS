```python
import heapq
```

```python
### Combining the new heirarchy with existing classes
# read the sample file having entities and the candidate classes
wiki_class_priority_dict = defaultdict(set)
wiki_class_priority_dict_list =defaultdict(list)

# wiki_class_dict_list
with open("candidate_classes.txt") as input_file:
    for index, line in enumerate(input_file):
        line= line.strip()
        if line:
            wiki_title, *classes= line.split(",")
            wiki_class_priority_dict[wiki_title]
            wiki_class_priority_dict_list[wiki_title]
            # proces the classes
            q = []
            for cls in classes:
                final_class = (cls.split("/")[-1])
                #read the class heirachy if exists
                if final_class in Dbpedia_priority:
#                     print(final_class,Dbpedia_priority[final_class],UNER_Dbpedia_v1[final_class])
                    heapq.heappush(q, (Dbpedia_priority[final_class], final_class))

                # get the hierarchy of the class
                if final_class in UNER_Dbpedia_v1 and UNER_Dbpedia_v1[final_class]:
                    wiki_class_priority_dict[wiki_title].add((UNER_Dbpedia_v1[final_class],Dbpedia_priority[final_class]))
                    wiki_class_priority_dict_list[wiki_title].append((UNER_Dbpedia_v1[final_class],Dbpedia_priority[final_class]))
        
```

```python
# how many empty entities in set wise analysis
class_count=[]
count=0
count_1 =0
count_2 =0
max_class_count =0
for i,j in wiki_class_priority_dict_list.items():
    class_count.append(len(j))
    if len(j)>max_class_count:
        max_class_count=len(j)
#     print(i,j)
    if len(j)==0:
        count+=1
    elif len(j)==1:
        count_1+=1
    elif len(j)==2:
        count_2+=1
print(count,count_1,count_2, max_class_count)
Counter(class_count)
```

```python
 # print the keys and values
#     json to analysed
final_json=[]
for key in wiki_class_priority_dict:
    q = []
    json_entity={}
    values = wiki_class_priority_dict[key]
#     if len(values)>2:
    for value in values:
        heapq.heappush(q, (value[1], value[0]))

    print(key)
    json_entity["title"]= key
    print(values)
    json_entity["candidates"] = values
    print(heapq.nlargest(1,q))
    json_entity["chosen_one"] =heapq.nlargest(1,q)
    print("\n")
    final_json.append(json_entity)
    
```

```python
class SetEncoder(json.JSONEncoder):
...    def default(self, obj):
...       if isinstance(obj, set):
...          return list(obj)
...       return json.JSONEncoder.default(self, obj)
```

```python
with open('wiki_class_priority_UNER_dict.json', 'w') as outfile:
    json.dump(final_json, outfile, cls=SetEncoder)
```
